## Implementación recursiva

````python
import numpy as np
import matplotlib.pyplot as plt



tiempos = []

def rec(base, exponente):
  if (exponente==0):
    return 1
  if (exponente==1):
    return base
  else:
    return base * rec(base,exponente-1)



valores = []

for i in range(1, 600):
    times=[]
    k=0
    while (k<=10):
      inicio= time.time()
      rec(2, i)
      fin = time.time()
      times.append(fin-inicio)
      k +=1
    valores.append(sum(times)/len(times))
````

### Complejidad teórica


```
if (exponente==0): -----------------------> 1
    return 1       -----------------------> 1
  if (exponente==1): ---------------------> 1
    return base      ---------------------> 1
  else:
    return base * rec(base,exponente-1) --> n(T(n-1))
```

$$
T(n) = nT(n-1) + 4
$$
$$
T(n) \in O(2^n)
$$



### Complejidad experimental
Se ejecutó en un tiempo promedio de 5.059772067599826e-06 segundos

````python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(8, 5)

ax.scatter(range(1, 600), valores, label="Tiempos", marker=".", color="blue")

x=np.linspace(0,200,600)
funcion_i="2**x"
y=lambda x: eval(funcion_i)
y=y(x)


tiempos_index= []
for i in range(1, len(y)+1):
  tiempos_index.append(i)
fig,ax=plt.subplots(1,1);fig.set_size_inches(11.758,5);
ax.scatter(tiempos_index, (y),label="", marker=".",color="red")

plt.show()
````

## Comparación entre el rendimiento del algoritmo recursivo vs divide y vencerás

El algoritmo implementado para la exponenciación recursiva, utiliza un enfoque iterativo simple, mientras que el segundo utiliza la
recursión y en adición, divide y vencerás.

### Complejidad temporal
- El algoritmo recursivo tiene una complejidad temporal lineal \(O(n)\) debido al bucle for que
itera sobre el exponente.
- El segundo código tiene una complejidad temporal logarítmica \(O(\log n)\) debido a la
naturaleza de la recursión y la estrategia de divide y vencerás.

### Rendimiento
En general, el algoritmo divide y vencerás es más eficiente para exponentes grandes, ya
que reduce el problema a la mitad en cada llamada recursiva.

## Divide y vencerás

#### COMPLEJIDAD TEORICA
$$
T(n) = 2*T\left(\frac{n}{2}\right) + O(n
)
$$
$$
T(n) \in O(n *log2(n))
$$

#### COMPLEJIDAD EXPERIMENTAL
La complejidad experimental toma un promedio de 1.2185480823778179e-05

````python
import time
import numpy as np
def div_V(num, exp):
  if exp == 0:
    return 1
  if exp == 1:
    return num
  else:
    if (exp % 2) == 0:
      a = div_V(num, exp//2)
      return a * a
    else:
      a = div_V(num, (exp-1)//2)
      return num * a * a

tiempos = []
for i in range(10000):
    times=[]
    k=0
    while (k<=30):
      inicio=time.time()
      div_V(2, i)
      fin=time.time()
      times.append(fin-inicio)
      k +=1
    tiempos.append(np.log(sum(times)/len(times)))
````

````python
tiempos_index= []
for i in range(1, len(tiempos)+1):
  tiempos_index.append(i)
fig,ax=plt.subplots(1,2);fig.set_size_inches(8,5);
ax[0].scatter(tiempos_index, (tiempos),label="Tiempos", marker=".",color="blue")
plt.ylim()
def nlogn(x):
    return x * np.log2(x)
x = np.linspace(1, 200, 10000)
y = nlogn(x)
ax[1].plot(x,np.log2(y))


plt.show()
````
